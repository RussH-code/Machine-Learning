## Removing Duplciates

```python
data = pd.DataFrame({'k1': ['one', 'two'] * 3 + ['two'],'k2': [1, 1, 2, 3, 3, 4, 4]})
data
```

```python
data.duplicated() #DataFrame method duplicated reutrns a boolean Series indicating whether each row is a duplicate or not
```

```python
data.drop_duplicates()
```

```python
data['v1'] = range(7)
data
```

We can filter duplciates based on a subset of columns

```python
data.drop_duplicates(['k1'])
```

- Duplicated and drop_duplicates by default keep the first observed value combination. Passing keep='last' will return the last one:

```python
data.drop_duplicates(['k1', 'k2'], keep='last')
```

# Transforming Data Using a Function or Mapping with __map()__

```python
data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
'Pastrami', 'corned beef', 'Bacon',
'pastrami', 'honey ham', 'nova lox'],
'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data
```

```python
meat_to_animal = {
'bacon': 'pig',
'pulled pork': 'pig',
'pastrami': 'cow',
'corned beef': 'cow',
'honey ham': 'pig',
'nova lox': 'salmon'
}
lowercased = data['food'].str.lower()
lowercased
```

```python
data['animal'] = lowercased.map(meat_to_animal) #amp() returns a Series with the same index as caller
data
```

# Discretization and Binning

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
```

```python
bins = [18, 25, 35, 60, 100]
cats = pd.cut(ages, bins)
cats
```

```python
print(cats.codes) #cut object has codes and categories attributes
print(cats.categories)
```

```python
pd.value_counts(cats)
```

```python
pd.cut(ages, [18, 26, 36, 61, 100], right=False) #right=False changes the right bracket from open to closed
```

```python
group_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
pd.cut(ages, bins, labels=group_names)
```

- If you pass an integer number of bins to cut instead of explicit bin edges, it
will compute equal-length bins based on the minimum and maximum values
in the data.

```python
data = np.random.rand(20)
pd.cut(data, 4, precision = 2)
```

- A closely related function, qcut, bins the data based on sample quantiles. With qcut, each bin will have the same amount of data points

```python
data = np.random.randn(1000)
cats = pd.qcut(data, 4) 
cats
```

```python
pd.value_counts(cats)
```

```python
pd.qcut(data, [0, 0.1, 0.5, 0.9, 1.]) #We can pass our own values as quantiles as well
```

# Computing Indicator/Dummy Variables

```python
df = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'b'],'data1': range(6)})
df
```

```python
pd.get_dummies(df['key'])
```

```python
import pandas as pd
dummies = pd.get_dummies(df['key'], prefix = 'key')
dummies
```

```python
df_with_dummies = df[['data1']].join(dummies)
df_with_dummies
```

```python
mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('https://raw.githubusercontent.com/wesm/pydata-book/2nd-edition/datasets/movielens/movies.dat', sep = '::', header=None, names = mnames)
```

```python
movies[:10]
```

```python
all_genres = []
for x in movies.genres:
    all_genres.extend(x.split('|'))
genres = pd.unique(all_genres)
genres
```

```python
import numpy as np
zero_matrix = np.zeros((len(movies), len(genres)))
dummies = pd.DataFrame(zero_matrix, columns=genres)
dummies
```

```python
for i, gen in enumerate(movies.genres):
    indices = dummies.columns.get_indexer(gen.split('|'))
    dummies.iloc[i, indices] = 1
dummies
```

```python
movies_windic = movies.join(dummies.add_prefix("Genres_"))
movies_windic
```
